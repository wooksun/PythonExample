Trident User Agents

https://developers.whatismybrowser.com/useragents/explore/layout_engine_name/trident/

```python
targetSite = 'https://www.melon.com/chart/index.htm'
header = {'User-agent': 'Mozilla/5.0 (Windows NT 6.3; WOW64; Trident/7.0; rv:11.0) like Gecko'}
request = requests.get(targetSite, headers=header)
print(request)
```

```python
html = request.text
soup = BeautifulSoup(html, 'html.parser')
#print(soup)
```

```python
# 노래 제목 크롤링
titles = soup.findAll('div', {'class': 'ellipsis rank01'})
#print(len(titles))
#print(titles)

for i in range(len(titles)):
    #print('{:3d}위 {}'.format(i + 1, titles[i].text.strip())) # 방법1
    #print('{:3d}위 {}'.format(i + 1, titles[i].text.split('\n')[1])) # 방법2
    print('{:3d}위 {}'.format(i + 1, titles[i].text.strip().split('\n')[0])) # 방법3
```

```python
# 아티스트 크롤링
artists = soup.findAll('div', {'class': 'ellipsis rank02'})
#print(len(titles))
#print(artists)

for i in range(len(artists)):
    #print(artists[i].text)
    #print('{:3d}위 {}'.format(i + 1, artists[i].text.strip())) # 적절하지 않음(두명 이상의 가수가 포함되면 진행불가)
    #print('{:3d}위 {}'.format(i + 1, artists[i].text.split('\n')[1])) # 적절하지 않음
    print('{:3d}위 {}'.format(i + 1, artists[i].text.strip().split('\n')[0]))
```

```python
# 아티스트 크롤링
artists = soup.findAll('span', {'class': 'checkEllipsis'})
#print(len(titles))
#print(artists)

for i in range(len(artists)):
    #print(artists[i].text)
    #print('{:3d}위 {}'.format(i + 1, artists[i].text.strip())) # 적절하지 않음(두명 이상의 가수가 포함되면 진행불가)
    #print('{:3d}위 {}'.format(i + 1, artists[i].text.split('\n')[1])) # 적절하지 않음
    print('{:3d}위 {}'.format(i + 1, artists[i].text.strip().split('\n')[0]))
```

```python
print(dt.now())
for i in range(100):
    artist = artists[i].text.strip().split('\n')[0]
    title = titles[i].text.strip().split('\n')[0]
    print('{:3d}위 {} - {}'.format(i + 1, artist, title))
```
