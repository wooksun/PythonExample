```python
import requests
from bs4 import BeautifulSoup
```

```python
targetSite = 'https://basicenglishspeaking.com/daily-english-conversation-topics/'
```

```python
request = requests.get(targetSite)
#print(request)
html = request.text
#print(html)
soup = BeautifulSoup(html, 'html.parser')
#print(soup)
```

```python
subjects = []
divs = soup.findAll('div', {'class': 'tcb-flex-col'})
# print(len(divs)) # 3
i = 0
for div in divs:
    #print(div)
    chapter = div.findAll('a')
    #print(len(chapter)) # 25
    for a in chapter:
        i += 1
        print('{:2d}. {}'.format(i, a.text))
        subjects.append(a.text)
        #print(subjects)
```

```python
for i in range(len(subjects)):
    print('{:2d}. {}'.format(i + 1, subjects[i]))
```
