```python
import warnings
warnings.filterwarnings(action='ignore')
%config Computer.use_jedi = False
import requests
from bs4 import BeautifulSoup
from datetime import datetime as dt
```

```python
print(dt.now())
```

```python
targetSite = 'https://music.bugs.co.kr/chart'
request = requests.get(targetSite)
#print(request)
html = request.text
#print(html)
soup = BeautifulSoup(html, 'html.parser')
#print(soup)
```

```python
# 노래 제목 크롤링
titles = soup.findAll('p', {'class': 'title'})
#print(len(titles))
#print(titles)

for i in range(len(titles)):
    #print('{:3d}위 {}'.format(i + 1, titles[i].text.strip())) # 방법1
    #print('{:3d}위 {}'.format(i + 1, titles[i].text.split('\n')[1])) # 방법2
    print('{:3d}위 {}'.format(i + 1, titles[i].text.strip().split('\n')[0])) # 방법3
```

```python
# 아티스트 크롤링
artists = soup.findAll('p', {'class': 'artist'})
#print(len(titles))
#print(artists)

for i in range(len(artists)):
    #print(artists[i].text)
    #print('{:3d}위 {}'.format(i + 1, artists[i].text.strip())) # 적절하지 않음(두명 이상의 가수가 포함되면 진행불가)
    #print('{:3d}위 {}'.format(i + 1, artists[i].text.split('\n')[1])) # 적절하지 않음
    print('{:3d}위 {}'.format(i + 1, artists[i].text.strip().split('\n')[0]))
```

```python
print(dt.now())
for i in range(100):
    artist = artists[i].text.strip().split('\n')[0]
    title = titles[i].text.strip().split('\n')[0]
    print('{:3d}위 {} - {}'.format(i + 1, artist, title))
```

```python
# 크롤링한 결과를 텍스트 파일에 저장한다.
#file = open('C:\\k_digital\\python\\crawling\\workspace\\output\\bugsMusic_05.txt', 'w') # 절대 경로
#file = open('C:/k_digital/python/crawling/workspace/output/bugsMusic_05.txt', 'w') # 절대 경로
file = open('./output/bugsMusic_05.txt', 'w', -1, 'utf-8') # 상대경로
file.write(dt.now().strftime('%Y-%m-%d %H:%M:%S') + '\n')
for i in range(100):
    artist = artists[i].text.strip().split('\n')[0]
    title = titles[i].text.strip().split('\n')[0]
    file.write('{:3d}위 {} - {}\n'.format(i + 1, artist, title))
file.close()
print('bugsMusic.txt 파일로 쓰기 완료')
```

```python
# with 구문을 사용해서 파일을 open하면 with 블록의 내용이 모두 실행된 후 자동으로 파일을 닫는다.
with open('./output/bugsMusic_05.txt', 'w', -1, 'utf-8') as file:
    file.write(dt.now().strftime('%Y-%m-%d %H:%M:%S') + '\n')
    for i in range(100):
        artist = artists[i].text.strip().split('\n')[0]
        title = titles[i].text.strip().split('\n')[0]
        file.write('{:3d}위 {} - {}\n'.format(i + 1, artist, title))
# ====== with
print('bugsMusic.txt 파일로 쓰기 완료')
```

```python
# 폴더에 저장된 파일 이름 가져오기
import os 
fileList = os.listdir('./output/')
print(fileList)
print(type(fileList))

for file in fileList:
    #print(file[:-4]) # 확장명이 무조건 3글자 일 때
    print(file[:file.rindex('.')]) # 파일 이름에 '.'이 여러개이고, 확장명 길이가 가변적일 때
```

```python
# 텍스트 파일에 저장된 데이터를 읽어서 화면에 출력한다.
try:
    file = open('./output/bugsMusic_05.txt', 'r', -1, 'utf-8')
    lines = file.readlines()
    for line in lines:
        print(line.strip())
except FileNotFoundError:
    print('디스크에 파일이 없습니다.')
```
